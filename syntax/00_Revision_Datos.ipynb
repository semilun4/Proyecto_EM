### Función para formatear los registros

```python
def formatear_registro(registros):
    list_reportes = []
    consecutivo = registros['Consecutivo']
    for idx_reporte, reporte in enumerate(registros['reportes']):
        data_reportes = {'Consecutivo': consecutivo, 'Numero reporte': idx_reporte}
        
        # Itera en los campos
        for k_campo, v_campo in reporte.items():
            if k_campo == 'Tipo Reporte':
                data_reportes['Tipo Reporte'] = v_campo         
            if isinstance(v_campo, list):
                v_campo = {i: v for i, v in enumerate(v_campo)}
            if isinstance(v_campo, dict):  
                data_reportes[k_campo] = dict()
                
                # Itera en subcampos
                for k_subcampo, v_subcampo in v_campo.items():
                    if isinstance(v_subcampo, list) and len(v_subcampo) == 1:
                        v_subcampo = v_subcampo[0]
                    subcampos = list()
                    try:
                        for inst, data_sub in v_subcampo.items():
                            data_sub = flatten_subcampo(data_sub)
                            data_sub['Institucion'] = inst
                            subcampos.append(data_sub)
                    except:
                        print('!', k_campo, k_subcampo, v_subcampo)
                    data_reportes[k_campo][k_subcampo] = subcampos
                    
        list_reportes.append(data_reportes)
    return list_reportes
```

```python
# Convierte las listas en diccionarios numerados
def dict_everything(data_structure):
    if type(data_structure) == list:
        data_structure = {str(i):v for i,v in enumerate(data_structure)}
    if type(data_structure) == dict:
        data_structure = {k:dict_everything(v) for k,v in data_structure.items()}
    return data_structure

# Aplana un diccionario anidado 
def flatten_subcampo(data, reverse_flatten_keys = True):
    def reverse(s, delimiter='_'):
        if type(s)==str and '_' in s:
            s = '_'.join( s.split('_')[::-1])
        return s
    data = dict_everything(data)
    data = flatdict.FlatDict(data, delimiter='_')
    if reverse_flatten_keys:
        data = {reverse(k):v for k,v in data.items()}
    return data
```

### Funciones tratamiento de registros duplicados

```python
# Función para combinar registros duplicados
def combinar_registros(grupo):
    registro_combinado = {}
    for columna in grupo.columns:
        # Filtrar valores no nulos y no "Desconocido"
        valores = grupo[columna].dropna()
        valores = valores[valores.apply(lambda x: str(x).strip().lower() != 'desconocido')]
        
        if not valores.empty:
            # Mantener el valor más frecuente
            valor_mas_frecuente = valores.mode().iloc[0]
            registro_combinado[columna] = valor_mas_frecuente
        else:
            # Asignar "Desconocido" si no hay valores válidos
            registro_combinado[columna] = 'Desconocido'
    
    return pd.Series(registro_combinado)

# Aplicar la función para combinar registros duplicados
def eliminar_duplicados(df, id_col):
    duplicados = df[df.duplicated(id_col, keep=False)]
    registros_combinados = duplicados.groupby(id_col).apply(combinar_registros)
    no_duplicados = df[~df[id_col].isin(duplicados[id_col])]
    # Concatenar los registros no duplicados con los registros combinados
    df_final = pd.concat([no_duplicados, registros_combinados]).reset_index(drop=True)
    return df_final
```

### Función principal de preprocesamiento

```python
def procesar_registros(registros, id_json):
    summary = []
    data_formateada = dict()
    
    for registro in registros:
        list_registros = formatear_registro(registro)  # list of dictionaries
        
        for reporte in list_registros:
            idx_consecutivo = str(reporte['Consecutivo']).zfill(6)
            idx_reporte = f"{id_json}-{idx_consecutivo}-{reporte['Numero reporte']}"
            data_formateada[idx_reporte] = reporte
            
            for k_campo, v_campo in reporte.items():
                if isinstance(v_campo, dict):
                    for k_subcampo, v_subcampo in v_campo.items():
                        for n_subcampo, data_sub in enumerate(v_subcampo):
                            #print(list(data_sub.keys()))
                            for key in data_sub.keys():
                                if key.startswith('Fecha de inicio del expediente'):
                                    fecha_reporte = data_sub[key]
                                    break
                                if key.startswith('Fecha de inicio del expediente'):
                                    fecha_reporte = data_sub[key]
                                    break
                            data_new = {
                                'IDX_REPORTE': idx_reporte,
                                'CAMPO': k_campo,
                                'SUBCAMPO': str(k_subcampo),
                                'N_SUBCAMPO': n_subcampo,
                                'INSTITUCION': data_sub['Institucion'],
                                'FECHA_REPORTE':  fecha_reporte
                            }
                            # Se agregan las columnas de conteo de valores al summary
                            data_new.update(contar_valores(data_sub))
                            
                            # Extraer y agregar otras variables a data_new
                            data_new['SEXO'] = extraer_sexo(data_sub)
                            data_new['EDAD'] = extraer_edad(data_sub)
                            data_new['FECHA_NACIMIENTO'] = extraer_fecha_nacimiento(data_sub)
                            data_new['ESTADO_CIVIL'] = extraer_estado_civil(data_sub)
                            data_new['NIVEL_EDUCATIVO'] = extraer_nivel_educativo(data_sub)
                            
                            summary.append(data_new)
    df = pd.DataFrame(summary)
    return df
```
